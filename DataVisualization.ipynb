# Query 1: Catagories the data with Fuel_type 

### 1.1 Bar Graph

```python
df.columns
```

```python
sns.catplot(x="Fuel_Type", kind="count", palette="ch:.25",edgecolor="0.6", data=df);
```

### Note Points:-
* Car with fuel type petrol is higher in numbers.
* Car with fuel type cng is lower in numbers.

### 1.2 pair plot based on Fuel Type

```python
g = sns.pairplot(df, hue="Fuel_Type", palette="Set2",diag_kind="hist", height=2.5)
```

### Note Points:-
* Owner 1 have most of the expensive cars.
* Mejority of the cars belongs to Owner 0.
